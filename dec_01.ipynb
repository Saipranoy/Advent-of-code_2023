```python
def text_file_to_array(file_path):
  with open(file_path,"r") as file:
    lines = file.readlines()
    arrays = [list(line.strip()) for line in lines]
  return arrays,lines
```

```python
file_path = "/content/input.txt"
result,lines = text_file_to_array(file_path)
```

```python
def calculate_sum_of_digits(input_str):
    # Extract numeric characters from the input string
    numeric_chars = [char for char in input_str if char.isdigit()]

    # Convert the list of numeric characters to integers and calculate the sum
    digit_sum = sum(map(int, numeric_chars))

    return digit_sum

# Test the function with the provided input
input_str = "two1nine"
output = calculate_sum_of_digits(input_str)
print(output)
```

```python
def text_to_num(a):
  c={}

  x = a.find("one")
  if x!= -1:
    c[x]=2

  x = a.find("two")
  if x!= -1:
    c[x]=2

  x = a.find("two")
  if x!= -1:
    c[x]=2

  x = a.find("three")
  if x!= -1:
    c[x]=3

  x = a.find("four")
  if x!= -1:
    c[x]=4

  x = a.find("five")
  if x!= -1:
    c[x]=5

  x = a.find("six")
  if x!= -1:
    c[x]=6

  x = a.find("seven")
  if x!= -1:
    c[x]=7

  x = a.find("eight")
  if x!= -1:
    c[x]=8

  x = a.find("nine")
  if x!= -1:
    c[x]=9
  b = []
  b.append(a)
  for i in range(len(b[0])):
    if b[0][i].isdigit():
      c[i]=int(b[0][i])

  sl = sorted(c.keys())

  y = c[sl[0]]*10+c[sl[len(sl)-1]]

  return y
```
